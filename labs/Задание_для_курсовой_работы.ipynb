–¢–µ–ø–µ—Ä—å –æ–Ω —Å–ø—Ä–∞—à–∏–≤–∞–µ—Ç –≤–∞—Å: –≤–æ–∑–º–æ–∂–Ω–æ –ª–∏ —Å–¥–µ–ª–∞—Ç—å –≤—Å–µ —ç–ª–µ–º–µ–Ω—Ç—ã –º–∞—Å—Å–∏–≤–∞ —Ä–∞–≤–Ω—ã–º–∏? –¢–∞–∫ –∫–∞–∫ –æ–Ω —Ö–æ—á–µ—Ç —Å–¥–µ–ª–∞—Ç—å –≤–∞—à—É –∂–∏–∑–Ω—å —Å–ª–æ–∂–Ω–µ–µ, —Ç–æ –≤–≤–æ–¥–∏—Ç –¥–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω–æ–µ –æ–≥—Ä–∞–Ω–∏—á–µ–Ω–∏–µ: –≤ –º–∞—Å—Å–∏–≤–µ –¥–æ–ª–∂–Ω–æ –æ—Å—Ç–∞—Ç—å—Å—è —Ö–æ—Ç—è –±—ã 2 —ç–ª–µ–º–µ–Ω—Ç–∞.

–í—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ
–ï—Å–ª–∏ –ï—Ö–∞–± –º–æ–∂–µ—Ç —Å–¥–µ–ª–∞—Ç—å –≤—Å–µ —ç–ª–µ–º–µ–Ω—Ç—ã –º–∞—Å—Å–∏–≤–∞ —Ä–∞–≤–Ω—ã–º–∏, –ø—Ä–∏–º–µ–Ω—è—è –æ–ø–∏—Å–∞–Ω–Ω—É—é –æ–ø–µ—Ä–∞—Ü–∏—é, —Ç–∞–∫, —á—Ç–æ–±—ã –≤ –º–∞—Å—Å–∏–≤–µ –æ—Å—Ç–∞–ª–æ—Å—å —Ö–æ—Ç—è –±—ã 2 —ç–ª–µ–º–µ–Ω—Ç–∞, –≤—ã–≤–µ–¥–∏—Ç–µ ¬´YES¬ª. –ò–Ω–∞—á–µ –≤—ã–≤–µ–¥–∏—Ç–µ ¬´NO¬ª.

–í–æ –≤—Ç–æ—Ä–æ–º –ø—Ä–∏–º–µ—Ä–µ –æ—Ç–≤–µ—Ç–∞ –Ω–µ —Å—É—â–µ—Å—Ç–≤—É–µ—Ç.

```python
### –ó–∞–¥–∞—á–∞ 1
#–ü—Ä–æ–≤–µ—Ä–∫–∞ —Å–ø–∏—Å–∫–∞
def zadanie1(arr):
    print("n=",n,"a=",arr)
    while len(arr) > 2:
        if all_the_same(arr):
            return True 
        xor_arr(arr)
    return all_the_same(arr)
#pre = []

#–ü—Ä–æ–≤–µ—Ä–∫–∞ —Ä–∞–≤–µ–Ω—Å—Ç–≤–∞ –≤—Å–µ—Ö —ç–ª–µ–º–µ–Ω—Ç–æ–≤
def all_the_same(elements):
    if len(elements) < 1:
        return True
    return len(elements) == elements.count(elements[0])

#–ò–∑–º–µ–Ω–µ–Ω–∏–µ —Å–ø–∏—Å–∫–∞ —Å –ø–æ–º–æ—â—å—é xor
def xor_arr(arr):
    for i in range(int(len(arr) / 2)):
        arr[i] = arr[i] ^ arr[i+1]
        del arr[i+1]
    return arr

#–ü—Ä–æ–≤–µ—Ä–∫–∞–¥–ª–∏–Ω—ã —Å–ø–∏—Å–∫–∞ –∏ —ç–ª–µ–º–µ–Ω—Ç–æ–≤ —Å–ø–∏—Å–∫–∞- 0‚â§ ùëéùëñ <230 
def test_a(n,a): 
    if n != len(a):
        print("–ù–µ–¥–æ–ø—É—Å—Ç–∏–º–∞—è –¥–ª–∏–Ω–∞, n,len()=",n,len(a))
        return False
    
    for ai in a:
        if ai < 0 or ai >= 230:
            print("–ù–µ–¥–æ–ø—É—Å—Ç–∏–º–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ a=",ai)
            return False
    return True

##########################################    
###–°—Ç–∞—Ä—Ç###    
filename="d:\\kurs_1.txt" #file with text
with open(filename) as file:
    line_counter=0
    t=0 #–∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –Ω–∞–±–æ—Ä–æ–≤ –¥–∞–Ω–Ω—ã—Ö
    for line in file:
        if line_counter==0: 
            t = int(line)
            #–ü—Ä–æ–≤–µ—Ä–∫–∞- —Ü–µ–ª–æ–µ —á–∏—Å–ª–æ ùë° (1‚â§ùë°‚â§15) ‚Äî –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –Ω–∞–±–æ—Ä–æ–≤ –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
            if t < 1 or t > 15:
                print("–ù–µ–¥–æ–ø—É—Å—Ç–∏–º–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ t=",t)
                break
            print("t=",t)
        else:    
            if t < 1: break #–≤—Å–µ –Ω–∞–±–æ—Ä—ã –æ–±—Ä–∞–±–æ—Ç–∞–Ω—ã,–∫–æ–Ω–µ—Ü
                
            #–æ–±—Ä–∞–±–æ—Ç–∫–∞ —Å—Ç—Ä–æ–∫ –Ω–∞–±–æ—Ä–æ–≤ –¥–∞–Ω–Ω—ã—Ö
            if (line_counter % 2) == 1:
                n = int(line)
                #–ü—Ä–æ–≤–µ—Ä–∫–∞- —Ü–µ–ª–æ–µ —á–∏—Å–ª–æ ùëõ (2‚â§ùëõ‚â§2000) ‚Äî —Ä–∞–∑–º–µ—Ä –º–∞—Å—Å–∏–≤–∞ ùëé.
                if n < 2 or n > 2000:
                    print("–ù–µ–¥–æ–ø—É—Å—Ç–∏–º–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ n=",n)
                    break
                
            else:
                a2 = list(line.split())
                a = list(map(int, a2))
                #–ü—Ä–æ–≤–µ—Ä–∫–∞- 0‚â§ ùëéùëñ <230
                if test_a(n,a)==False: break
                    
                #–≤—ã–∑—ã–≤–∞–µ–º –≥–ª–∞–≤–Ω—É—é —Ñ—É–Ω–∫—Ü–∏—é –¥–ª—è –ø–æ–¥—Å—á–µ—Ç–∞
                print('Yes' if zadanie1(a) else 'No')
                t = t-1
        line_counter=line_counter+1    
```

### –ó–∞–¥–∞—á–∞ 2
–®–∫–æ–ª—å–Ω–∏–∫ –í–∞—Å—è –ø–æ–ª—É—á–∏–ª –≤ –ø–æ–¥–∞—Ä–æ–∫ –Ω–∞ –¥–µ–Ω—å —Ä–æ–∂–¥–µ–Ω–∏—è —Å—Ç—Ä–æ–∫—É –¥–ª–∏–Ω—ã n, —Å–æ—Å—Ç–æ—è—â—É—é –∏–∑ –±—É–∫–≤ ¬´a¬ª –∏ ¬´b¬ª. –í–∞—Å—è –Ω–∞–∑—ã–≤–∞–µ—Ç –ø—Ä–∏–≤–ª–µ–∫–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—å—é —Å—Ç—Ä–æ–∫–∏ –º–∞–∫—Å–∏–º–∞–ª—å–Ω—É—é –¥–ª–∏–Ω—É –ø–æ–¥—Å—Ç—Ä–æ–∫–∏ (–ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç–∏ —Å–æ—Å–µ–¥–Ω–∏—Ö —Å–∏–º–≤–æ–ª–æ–≤), —Å–æ—Å—Ç–æ—è—â–µ–π –∏–∑ –æ–¥–∏–Ω–∞–∫–æ–≤—ã—Ö —Å–∏–º–≤–æ–ª–æ–≤.

–í–∞—Å—è –º–æ–∂–µ—Ç –ø–æ–º–µ–Ω—è—Ç—å –≤ –∏—Å—Ö–æ–¥–Ω–æ–π —Å—Ç—Ä–æ–∫–µ –Ω–µ –±–æ–ª–µ–µ k —Å–∏–º–≤–æ–ª–æ–≤. –ö–∞–∫–æ–π –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–π –ø—Ä–∏–≤–ª–µ–∫–∞—Ç–µ–ª—å–Ω–æ—Å—Ç–∏ –¥–∞–Ω–Ω–æ–π —Å—Ç—Ä–æ–∫–∏ –æ–Ω —Å–º–æ–∂–µ—Ç –¥–æ–±–∏—Ç—å—Å—è?

–í—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ\
–í –ø–µ—Ä–≤–æ–π —Å—Ç—Ä–æ–∫–µ –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö —Å–æ–¥–µ—Ä–∂–∞—Ç—Å—è –¥–≤–∞ —á–∏—Å–ª–∞ n –∏ k (1‚Äâ‚â§‚Äân‚Äâ‚â§‚Äâ100‚Äâ000,‚Äâ0‚Äâ‚â§‚Äâk‚Äâ‚â§‚Äân) ‚Äî –¥–ª–∏–Ω–∞ —Å—Ç—Ä–æ–∫–∏ –í–∞—Å–∏ –∏ –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–µ —Ä–∞–∑—Ä–µ—à—ë–Ω–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∏–∑–º–µ–Ω–µ–Ω–∏–π.

–í–æ –≤—Ç–æ—Ä–æ–π —Å—Ç—Ä–æ–∫–µ —Å–ª–µ–¥—É–µ—Ç —Å—Ç—Ä–æ–∫–∞ –í–∞—Å–∏, —Å–æ—Å—Ç–æ—è—â–∞—è –∏–∑ –±—É–∫–≤ ¬´a¬ª –∏ ¬´b¬ª.

–í–æ –≤—Ç–æ—Ä–æ–º –ø—Ä–∏–º–µ—Ä–µ –æ–ø—Ç–∏–º–∞–ª—å–Ω—ã–π –æ—Ç–≤–µ—Ç –¥–æ—Å—Ç–∏–≥–∞–µ—Ç—Å—è –Ω–∞ —Å—Ç—Ä–æ–∫–µ ¬´aaaaabaa¬ª –∏–ª–∏ –Ω–∞ —Å—Ç—Ä–æ–∫–µ ¬´aabaaaaa¬ª.

```python
### –ó–∞–¥–∞—á–∞ 2
#n-–¥–ª–∏–Ω–∞ —Å—Ç—Ä–æ–∫–∏
#k-—Ä–∞–∑—Ä–µ—à–µ–Ω–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∑–∞–º–µ–Ω
def zadanie2(n,k,s):
    print("n=",n,"k=",k ,"s=",s)
    #–í—Ä–µ–º–µ–Ω–Ω—ã–µ –ø–µ—Ä–µ–º–µ–Ω–Ω—ã–µ
    c=0 #—Ä–µ–∑—É–ª—å—Ç–∞—Ç, –≤—Å–µ–≥–æ –ø—Ä–∏–≤–ª–µ–∫–∞—Ç–µ–ª—å–Ω—ã—Ö –±—É–∫–≤
    a=0 #—Å—á–µ—Ç—á–∏–∫ –ø—Ä–∏–≤–ª–µ–∫–∞—Ç–µ–ª—å–Ω—ã—Ö –∞   
    b=0 #—Å—á–µ—Ç—á–∏–∫ –ø—Ä–∏–≤–ª–µ–∫–∞—Ç–µ–ª—å–Ω—ã—Ö b
    t=0 #–∏–Ω–¥–µ–∫—Å –±—É–∫–≤—ã –∫–æ–≥–¥–∞ –Ω–µ –¥–æ–±–∞–≤–ª—è–µ–º –±—É–∫–≤—É –∫ —Å—á–µ—Ç—á–∏–∫—É –∞ –∏–ª–∏ b 
    for i in range(0, n):
        #—Å—á–∏—Ç–∞–µ–º –±—É–∫–≤—ã –∞ –∏ b 
        if s[i] == 'a': a=a+1
        else: b=b+1
        #print("i=",i, " a=",a, " b=",b," c=",c)
        #–µ—Å–ª–∏ –º–∏–Ω–∏–º–∞–ª—å–Ω–æ –≤—Å—Ç—Ä–µ—á–∞—é—â–∞—è—Å—è –±—É–∫–≤–∞ <= k(–∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∑–∞–º–µ–Ω)
        if min(a , b) <= k:
            #–º–æ–∂–µ–º –æ–±—å–µ–¥–∏–Ω–∏—Ç—å –º–∞–∫—Å –≤—Å—Ç—Ä–µ—Ç–∏–≤—à–∏–µ—Å—è –±—É–∫–≤—ã —Å k –º–∏–Ω–∏–º–∞–ª—å–Ω–æ –≤—Å—Ç—Ä–µ—Ç–∏–≤—à–∏–º–∏—Å—è –±—É–∫–≤–∞–º–∏
            c = max(c , a + b)
            #print("i=",i, " a=",a, " b=",b, " a+b=",a+b, " c=",c)
        else:
            #–Ω–µ –º–æ–∂–µ–º –∑–∞–º–µ–Ω–∏—Ç—å –≤—Å–µ –º–∏–Ω–∏–º–∞–ª—å–Ω–æ –≤—Å—Ç—Ä–µ—á–∞—é—â–∞—è—Å—è –±—É–∫–≤—ã
            #–ø–æ—ç—Ç–æ–º—É –Ω–∞–¥–æ –≤–æ—Å—Å—Ç–∞–Ω–æ–≤–∏—Ç—å –Ω—É–∂–Ω—ã–π —Å—á–µ—Ç—á–∏–∫ –±—É–∫–≤
            #–ø–æ—Å–∫–æ–ª—å–∫—É —ç—Ç—É –±—É–∫–≤—É –Ω–µ–ª—å–∑—è –æ–±—å–µ–¥–∏–Ω–∏—Ç—å –≤ –≥—Ä—É–ø–ø—É –æ–¥–∏–Ω–∞–∫–æ–≤—ã—Ö –±—É–∫–≤
            if s[t] == 'a': a=a-1
            else: b=b-1
            #print("*i=",i,"*s(",t,")=",s[t], "a=",a ,"b=",b, " c=",c)
            t=t+1    
    
    print(c)
    

#–ü—Ä–æ–≤–µ—Ä–∫–∞ —Å—Ç—Ä–æ–∫–∏ –í–∞—Å–∏, —Å–æ—Å—Ç–æ—è—â–∞—è –∏–∑ –±—É–∫–≤ ¬´a¬ª –∏ ¬´b¬ª
def test_a(n,a): 
    if n != len(a):
        print("–ù–µ–¥–æ–ø—É—Å—Ç–∏–º–∞—è –¥–ª–∏–Ω–∞, n,len()=",n,len(a))
        return False
    
    for ai in a:
        if ai == 'a' or ai == 'b':
            continue
        else:
            print("–ù–µ–¥–æ–ø—É—Å—Ç–∏–º–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ s=",ai)
            return False
    return True

##########################################    
###–°—Ç–∞—Ä—Ç###
filename="d:\\kurs_2.txt" #file with text
with open(filename) as file:
    line_counter=1
    for line in file:
        #–æ–±—Ä–∞–±–æ—Ç–∫–∞ —Å—Ç—Ä–æ–∫ –Ω–∞–±–æ—Ä–æ–≤ –¥–∞–Ω–Ω—ã—Ö
        if (line_counter % 2) == 1:
            tmp = line.split()
            n = int(tmp[0])
            k = int(tmp[1])
            
            #–ü—Ä–æ–≤–µ—Ä–∫–∞- 1‚Äâ‚â§‚Äân‚Äâ‚â§‚Äâ100‚Äâ000
            if n < 1 or n > 100000:
                print("–ù–µ–¥–æ–ø—É—Å—Ç–∏–º–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ n=",n)
                break  
            #–ü—Ä–æ–≤–µ—Ä–∫–∞- 0‚Äâ‚â§‚Äâk‚Äâ‚â§‚Äân
            if k < 0 or k > n:
                print("–ù–µ–¥–æ–ø—É—Å—Ç–∏–º–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ k=",k)
                break            
                 
        else:
            s = line.replace('\n',"")
            #–ü—Ä–æ–≤–µ—Ä–∫–∞ —Å—Ç—Ä–æ–∫–∞ —Å–æ–¥–µ—Ä–∂–∏—Ç ¬´a¬ª –∏ ¬´b¬ª –∏ –ø—Ä–∞–≤–∏–ª—å–Ω–æ–π –¥–ª–∏–Ω—ã
            if test_a(n,s)==False: break
            
            #–≤—ã–∑—ã–≤–∞–µ–º —Ñ—É–Ω–∫—Ü–∏—é –¥–ª—è –ø–æ–¥—Å—á–µ—Ç–∞
            zadanie2(n,k,s) 
        line_counter=line_counter+1           
```

–µ—Å–ª–∏ ùëô=909, —Ç–æ –ø—Ä–∏–±–∞–≤–ª–µ–Ω–∏–µ –µ–¥–∏–Ω–∏—Ü—ã –ø—Ä–∏–≤–µ–¥—ë—Ç –∫ —Ä–µ–∑—É–ª—å—Ç–∞—Ç—É 910 –∏ –±—É–¥—É—Ç –∏–∑–º–µ–Ω–µ–Ω—ã 2 —Ü–∏—Ñ—Ä—ã;\

```python
#–ó–∞–¥–∞—á–∞ 3
#–ü–æ–¥—Å—á–∏—Ç–∞—Ç—å —Ä–∞–∑–Ω—ã–µ —Ü–∏—Ñ—Ä—ã –≤ –¥–≤—É—Ö —á–∏—Å–ª–∞—Ö
#–ó–∞–¥–∞–Ω—ã –¥–≤–∞ —Ü–µ–ª—ã—Ö —á–∏—Å–ª–∞ ùëô –∏ ùëü, –≥–¥–µ ùëô<ùëü. –ë—É–¥–µ–º –ø—Ä–∏–±–∞–≤–ª—è—Ç—å 1 –∫ ùëô –¥–æ —Ç–µ—Ö –ø–æ—Ä, –ø–æ–∫–∞ —Ä–µ–∑—É–ª—å—Ç–∞—Ç –Ω–µ –æ–∫–∞–∂–µ—Ç—Å—è —Ä–∞–≤–Ω—ã–º ùëü
def countDiff(L,R):
    cnt = 0
    while L != 0 or R != 0:
        cnt += R - L
        L //= 10
        R //= 10
    print(cnt)

##########################################    
###–°—Ç–∞—Ä—Ç###
filename="d:\\kurs_3.txt" #file with text
with open(filename) as file:
    line_counter=0
    t=0 #–∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –Ω–∞–±–æ—Ä–æ–≤ –¥–∞–Ω–Ω—ã—Ö
    for line in file:
        if line_counter==0: 
            t = int(line)
            #–ü—Ä–æ–≤–µ—Ä–∫–∞- (1‚â§ùë°‚â§10^4) ‚Äî –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –Ω–∞–±–æ—Ä–æ–≤ –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
            if t < 1 or t > 10000:
                print("–ù–µ–¥–æ–ø—É—Å—Ç–∏–º–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ t=",t)
                break
            print("t=",t)
        else:    
            if t < 1: break #–≤—Å–µ –Ω–∞–±–æ—Ä—ã –æ–±—Ä–∞–±–æ—Ç–∞–Ω—ã,–∫–æ–Ω–µ—Ü
            #–æ–±—Ä–∞–±–æ—Ç–∫–∞ —Å—Ç—Ä–æ–∫ –Ω–∞–±–æ—Ä–æ–≤ –¥–∞–Ω–Ω—ã—Ö
            tmp = line.split()
            l = int(tmp[0])
            r = int(tmp[1])
            
            #–ü—Ä–æ–≤–µ—Ä–∫–∞- 1‚â§ ùëô < ùëü ‚â§10^9
            if l < 1 or l >= r or r >1e+9:
                print("–ù–µ–¥–æ–ø—É—Å—Ç–∏–º–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ l=",l," r=",r)
                break
           
                
            print("l=",l,"r=",r)
            countDiff(l,r)    
        line_counter=line_counter+1   
```

```python
#–ó–∞–¥–∞—á–∞ 7
#–í—ã–≤–µ–¥–∏—Ç–µ –µ–¥–∏–Ω—Å—Ç–≤–µ–Ω–Ω–æ–µ —Ü–µ–ª–æ–µ —á–∏—Å–ª–æ 0‚â§ùë•<109+7, –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Ü–∏–∫–ª–∏—á–µ—Å–∫–∏—Ö –ø–µ—Ä–µ—Å—Ç–∞–Ω–æ–≤–æ–∫ –¥–ª–∏–Ω—ã ùëõ –ø–æ –º–æ–¥—É–ª—é 109+7.
#–í—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ- –æ–¥–Ω–æ —Ü–µ–ª–æ–µ —á–∏—Å–ª–æ ùëõ (3‚â§ùëõ‚â§10^6).
def zadanie7(n):
    print("n=",n)
    MOD = int(1e9+7)
    print(MOD)
    res = 1
    fact = 1
    for i in range(1, n):
        
        res *= 2
        fact *= i
        fact = fact % MOD
        res = res % MOD
    
    fact *= n
    fact = fact % MOD
    fact -= res
    fact = fact % MOD
    if fact < 0: fact = fact + MOD
    print(fact) 

##########################################    
###–°—Ç–∞—Ä—Ç###
input = [4, 106]
for n in input: 
    #–ü—Ä–æ–≤–µ—Ä–∫–∞- 3‚â§ùëõ‚â§10^6
    if n < 3 or n > 1e+6:
        print("–ù–µ–¥–æ–ø—É—Å—Ç–∏–º–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ n=",n)
        break 
    zadanie7(n) 
```

```python
#–ó–∞–¥–∞—á–∞ 9
#–ù–∞–π–¥–∏—Ç–µ –º–∏–Ω–∏–º–∞–ª—å–Ω—É—é —Å—Ç–æ–∏–º–æ—Å—Ç—å, –Ω–µ–æ–±—Ö–æ–¥–∏–º—É—é –¥–ª—è –ø—Ä–µ–≤—Ä–∞—â–µ–Ω–∏—è ùëé0,ùëé1,...,ùëéùëõ‚àí1 –≤ —Å—Ç–µ–ø–µ–Ω–Ω—É—é –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—å.
#–í—Ö–æ–¥: –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—å –∏–∑ ùëõ –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã—Ö —á–∏—Å–µ–ª ùëé0,ùëé1,...,ùëéùëõ‚àí1
def zadanie9(n,a):
    inf = 10**18
    if n <= 2:
        print(a[0] - 1)
    else:
        result = sum(a) - n

        for x in range(1, 10**9):
            curPow = 1
            curCost = 0
            for i in range(n):
                curCost += abs(a[i] - curPow)
                curPow *= x
                if curPow > inf: break

            if curPow > inf: break
            if curPow / x > result + a[n - 1]: break
            result = min(result, curCost)

        print(result)

#–ü—Ä–æ–≤–µ—Ä–∫–∞ –¥–ª–∏–Ω—ã —Å–ø–∏—Å–∫–∞ –∏ —ç–ª–µ–º–µ–Ω—Ç–æ–≤ —Å–ø–∏—Å–∫–∞- 1‚â§ ùëéùëñ ‚â§10^9 
def test_a(n,a): 
    if n != len(a):
        print("–ù–µ–¥–æ–ø—É—Å—Ç–∏–º–∞—è –¥–ª–∏–Ω–∞, n,len()=",n,len(a))
        return False
    
    #–ü—Ä–æ–≤–µ—Ä–∫–∞ —ç–ª–µ–º–µ–Ω—Ç–∞ 1‚â§ùëéùëñ‚â§10^9
    for ai in a:
        if ai < 1 or ai > 1e+9:
            print("–ù–µ–¥–æ–ø—É—Å—Ç–∏–º–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ a=",ai)
            return False
    return True

##########################################    
###–°—Ç–∞—Ä—Ç###
filename="d:\\kurs_9.txt" #file with text
with open(filename) as file:
    line_counter=0
    for line in file:
        if line_counter%2==0: 
            n = int(line)
            #–ü—Ä–æ–≤–µ—Ä–∫–∞- 3‚â§ ùëõ ‚â§10^5
            if n < 3 or n > 1e+5: 
                print("–ù–µ–¥–æ–ø—É—Å—Ç–∏–º–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ n=",n)
                break
        else: 
            a2 = list(line.split())
            a = list(map(int, a2))
            
            #–ü—Ä–æ–≤–µ—Ä–∫–∞- 0‚â§ùëéùëñ‚â§1
            if test_a(n,a)==False: break 
                    
            #–≤—ã–∑—ã–≤–∞–µ–º –≥–ª–∞–≤–Ω—É—é —Ñ—É–Ω–∫—Ü–∏—é –¥–ª—è –ø–æ–¥—Å—á–µ—Ç–∞
            print("n=",n,"a=",a)
            a.sort()
            zadanie9(n,a)
            
        line_counter=line_counter+1          
```

```python
#–ó–∞–¥–∞—á–∞ 10
counter_n =0 #—Å—É–º–º–∞ ùëõ –ø–æ –≤—Å–µ–º –Ω–∞–±–æ—Ä–∞–º

#–í—Ö–æ–¥–Ω—ã–µ –ø–∞—Ä–∞–º–µ—Ç—Ä—ã
#n-–∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –±–æ—Å—Å–æ–≤
#ai-—Ç–∏–ø—É ùëñ-–≥–æ –±–æ—Å—Å–∞.0-–ª–µ–≥–∫–∏–π,1-—Å–ª–æ–∂–Ω—ã–π
def zadanie10(n,a):
    print("n=",n,"a=",a)
    #–ø–µ—Ä–≤–æ–≥–æ –±–æ—Å—Å–∞ –æ–±—Ä–∞–±–∞—Ç—ã–≤–∞–µ–º –æ—Ç–¥–µ–ª—å–Ω–æ
    #–µ—Å–ª–∏=1 —Ç–æ –¥–æ–±–∞–≤–∏–º 1 —á—Ç–æ–±—ã —É–±–∏—Ç—å –µ–≥–æ
    rez = a[0]; 
    i=0
    while True: 
        i = i+1
        if i >= n: break
        #print("#1 i=",i)
        if a[i] == 0: continue
        j = i

        while (j < n and a[j] == 1): j=j+1
        #–µ—Å–ª–∏ –¥–ª–∏–Ω–∞ —Ç–µ–∫—É—â–µ–≥–æ –æ—Ç—Ä–µ–∑–∫–∞ —Å–ª–æ–∂–Ω—ã—Ö –±–æ—Å—Å–æ–≤ —Ä–∞–≤–Ω–∞ k, 
        #—Ç–æ –Ω—É–∂–Ω–æ k/3 –æ—á–∫–æ–≤ –ø—Ä–æ–ø—É—Å–∫–∞.–ê–∫–∫—É–º—É–ª–∏—Ä—É–µ–º –ø–æ –≤—Å–µ–º –æ—Ç—Ä–µ–∑–∫–∞–º
        rez += (j - i) // 3
        i = j - 1
        #print("#2 i=",i)
    print(rez)
    print()

#–ü—Ä–æ–≤–µ—Ä–∫–∞ —ç–ª–µ–º–µ–Ω—Ç–æ–≤ —Å–ø–∏—Å–∫–∞- 0‚â§ ùëéùëñ ‚â§1
def test_a(n,a):   
    if n != len(a):
        print("–ù–µ–¥–æ–ø—É—Å—Ç–∏–º–∞—è –¥–ª–∏–Ω–∞, n,len()=",n,len(a))
        return False
    
    for ai in a:
        if ai == 0 or ai == 1:
            continue
        else:
            print("–ù–µ–¥–æ–ø—É—Å—Ç–∏–º–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ a=",ai)
            return False
    return True

##########################################    
###–°—Ç–∞—Ä—Ç###    
filename="d:\\kurs_10.txt" #file with text
with open(filename) as file:
    line_counter=0
    t=0 #–∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –Ω–∞–±–æ—Ä–æ–≤ –¥–∞–Ω–Ω—ã—Ö
    for line in file:
        if line_counter==0: 
            t = int(line)
            #–ü—Ä–æ–≤–µ—Ä–∫–∞- 1‚â§ ùë° ‚â§2‚ãÖ10^4
            if t < 1 or t > 2e+4:
                print("–ù–µ–¥–æ–ø—É—Å—Ç–∏–º–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ t=",t)
                break 
            print("t=",t)
        else:    
            if t < 1: break #–≤—Å–µ –Ω–∞–±–æ—Ä—ã –æ–±—Ä–∞–±–æ—Ç–∞–Ω—ã,–∫–æ–Ω–µ—Ü
                
            #–æ–±—Ä–∞–±–æ—Ç–∫–∞ —Å—Ç—Ä–æ–∫ –Ω–∞–±–æ—Ä–æ–≤ –¥–∞–Ω–Ω—ã—Ö
            if (line_counter % 2) == 1:
                n = int(line)
                
                #–ü—Ä–æ–≤–µ—Ä–∫–∞- 1‚â§ ùëõ ‚â§2‚ãÖ10^5
                if n < 1 or n > 2e+5:
                    print("–ù–µ–¥–æ–ø—É—Å—Ç–∏–º–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ n=",n)
                    break 
                    
                #–ü—Ä–æ–≤–µ—Ä–∫–∞- —Å—É–º–º–∞ ùëõ –ø–æ –≤—Å–µ–º –Ω–∞–±–æ—Ä–∞–º –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö –Ω–µ –ø—Ä–µ–≤–æ—Å—Ö–æ–¥–∏—Ç 3‚ãÖ10^5.
                counter_n = counter_n + n
                #print("counter_n",counter_n)
                if counter_n > 2e+5:
                    print("–ù–µ–¥–æ–ø—É—Å—Ç–∏–º–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ —Å—É–º–º–∞ ùëõ,n=",counter_n)
                    break                     
                    
            else:
                a2 = list(line.split())
                a = list(map(int, a2))
                #–ü—Ä–æ–≤–µ—Ä–∫–∞- 0‚â§ùëéùëñ‚â§1
                if test_a(n,a)==False: break 
                    
                #–≤—ã–∑—ã–≤–∞–µ–º –≥–ª–∞–≤–Ω—É—é —Ñ—É–Ω–∫—Ü–∏—é –¥–ª—è –ø–æ–¥—Å—á–µ—Ç–∞
                zadanie10(n,a)
                t = t-1
        line_counter=line_counter+1           
```
