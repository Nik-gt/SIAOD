### Задание №1
Отсортировать строки файла, содержащие названия книг, в алфавитном порядке с использованием двух деков.

```python
#https://dontsova-books.ru/page/viola-tarakanova-spisok-knig-po-porjadku
###########################
# Реализация двухсторонней очереди в Python
class Deque:
    def __init__(self):
        self.items = []

    def isEmpty(self):
        return self.items == []

    def addRear(self, item):
        self.items.append(item)

    def addFront(self, item):
        self.items.insert(0, item)

    def removeFront(self):
        return self.items.pop()

    def removeRear(self):
        return self.items.pop(0)

    def size(self):
        return len(self.items)

##################################
#start    
deque = Deque()
deque2 = Deque()

#Возвращает min значение из deque
#При этом данные копируются вынужденно deque-->deque2
def findMinimum():
    if deque.isEmpty()==True: return ""
    
    min = deque.removeRear()
    if deque.isEmpty()==True: return min
    
    #find min
    while deque.isEmpty()==False: 
        current = deque.removeRear()
        if current < min:
            deque2.addFront(min) 
            min = current
        else: 
            deque2.addFront(current) 
    return min
##################################

filename="d:\\demo4_1.txt" #good file

#Сохраняем строки в deque
with open(filename) as file:
    for line in file:
        deque.addFront(line)
        print(line) 
        
print("--------------sorted")
#Ищем и печатаем минимум в deque, при этом данные переносятся в deque2
print(findMinimum())        

#Перенесем данные обратно в deque
while True:        
    # copy deque2-->deque      
    while deque2.isEmpty()==False: deque.addFront(deque2.removeRear())   
    print(findMinimum())
    if findMinimum() != "": break

print("ok")
```

### Задание №3
Даны три стержня и n дисков различного размера. Диски можно надевать на стержни, образуя из них башни. Перенести n дисков со стержня А на стержень С, сохранив их первоначальный порядок. При переносе дисков необходимо соблюдать следующие правила:
- на каждом шаге со стержня на стержень переносить только один диск;
- диск нельзя помещать на диск меньшего размера;
- для промежуточного хранения можно использовать стержень В.
Реализовать алгоритм, используя три стека вместо стержней А, В, С. Информация о дисках хранится в исходном файле.

```python
###########################
# Реализация стека в python
# Создаем стек
def create_stack():
    stack = []
    return stack

#  стек пустой?
def check_empty(stack):
    return len(stack) == 0

# Добавляем элементы в стек
def push(stack, item):
    stack.append(item)
    #print("Добавлен элемент: " + item)

# Удаляем элементы из стека
def pop(stack):
    if (check_empty(stack)):
        return "Стек пуст"
    return stack.pop()

def size(stack):
    return len(stack)

##################################
#Главная функция решения задачи
def tower_of_hanoi(numbers, start, auxiliary, result):
    i = 1
    start = create_stack()
    auxiliary = create_stack()
    result = create_stack()
    #загрузим в дек
    for j in numbers: push(start, j)
    s = size(start)
    direction = True # A to C 
    while size(result) != s:
        if check_empty(start) == True: direction = False # C to A
        if check_empty(auxiliary) == True: direction = True # A to C
        if direction ==True:
            element = pop(start) #удаляем из А
            if element == i: 
                push(result, element)
                print(start, result, auxiliary)
                i += 1
                continue
            push(auxiliary, element) #добавляем в С
            continue
        element = pop(auxiliary) #удаляем из C
        if element == i: 
            push(result, element)
            print(start, result, auxiliary)
            i += 1
            continue
        push(start, element) #добавляем в A

#Начало исполнения кода  
numbers = [3, 1, 2, 5, 4]
tower_of_hanoi(numbers, 'A', 'B', 'C')
```

### Задание №6
Дан файл из символов. Используя стек, за один просмотр файла напечатать сначала все цифры, затем все буквы, и, наконец, все остальные символы, сохраняя исходный порядок в каждой группе символов.

```python
###########################
# Реализация стека в python
# Создаем стек
def create_stack():
    stack = []
    return stack

#  стек пустой?
def check_empty(stack):
    return len(stack) == 0

# Добавляем элементы в стек
def push(stack, item):
    stack.append(item)
    #print("Добавлен элемент: " + item)

# Удаляем элементы из стека
def pop(stack):
    if (check_empty(stack)):
        return "Стек пуст"
    return stack.pop()

##################################
##Начало исполнения кода
stack=create_stack() #stack цифры
stack2=create_stack() #stack буквы
stack3=create_stack() #stack остальные символы
stackReverse=create_stack() #stack цифры


filename="d:\\demo4_6.txt" #good file
filename="d:\\demo4_6simple.txt" #good file


with open(filename) as file:
    for line in file:
        ss = list(line) #string to char array 
        for s1 in ss: 
            if s1.isdigit(): push(stack, s1)
            elif s1.isalpha(): push(stack2, s1)
            elif s1 == ' ': pass
            elif s1 == '\n': pass
            else: push(stack3, s1)
                
print("цифры")
while check_empty(stack)==False: push(stackReverse, pop(stack))    
#print    
while check_empty(stackReverse)==False: print(pop(stackReverse))   
    
print("буквы")
while check_empty(stack2)==False: push(stackReverse, pop(stack2))    
#print    
while check_empty(stackReverse)==False: print(pop(stackReverse))     
    
print("остальные символы")
while check_empty(stack3)==False: push(stackReverse, pop(stack3))    
#print    
while check_empty(stackReverse)==False: print(pop(stackReverse))     
```

### Задание №8
Дан текстовый файл. Используя стек, сформировать новый текстовый файл, содержащий строки исходного файла, записанные в обратном порядке: первая строка становится последней, вторая – предпоследней и т.д.

```python
###########################
# Реализация стека в python
# Создаем стек
def create_stack():
    stack = []
    return stack

#  стек пустой?
def check_empty(stack):
    return len(stack) == 0

# Добавляем элементы в стек
def push(stack, item):
    stack.append(item)
    #print("Добавлен элемент: " + item)

# Удаляем элементы из стека
def pop(stack):
    if (check_empty(stack)):
        return "Стек пуст"
    return stack.pop()

##################################
#start
stack=create_stack() #create stack

filename="d:\\demo4_8.txt" #good file

with open(filename) as file:
    for line in file: 
        push(stack, line)
        
while check_empty(stack)==False: print(pop(stack))            
```
